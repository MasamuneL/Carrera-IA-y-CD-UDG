# Ejercicio de Práctica:
Simular un sistema de impresión con varios documentos, en el que **solo 2 impresoras están disponibles**. 10 hilos deben representar a usuarios intentando imprimir.
### TAREA:
- Implementar la versión con un Semaphore que permita 2 impresiones simultáneas.
- Añadir sleeps pequeños para simular impresión.
- En una sección aparte, u otro archivo .py, modificar el código para mostrar qué ocurre si no se usa semáforo.

```python
# Importemos las librerías
import threading
import time
```

```python
# Creo un semaforo que simule que solo hay 2 impresoras

semaforo = threading.Semaphore(2)

# funcion que simula la impresion

def mandar_imprimr(user):
    with semaforo:
        print(f'El usuario {user} mando a imprimir')
        time.sleep(1)
        print(f'Se imprimio el documento del usuario {user}\n')

#simulamos que cada hilo es un cliente

hilos = []

for i in range(10):
    t = threading.Thread(target=mandar_imprimr, args=(i,))
    t.start()
    hilos.append(t)

for t in hilos:
    t.join()

print('Operaciones terminadas')
```

### Ejemplo sin semaforo

```python
# funcion que simula la impresion

def mandar_imprimr(user):
    print(f'El usuario {user} mando a imprimir')
    time.sleep(1)
    print(f'Se imprimio el documento del usuario {user}\n')

#simulamos que cada hilo es un cliente

hilos = []

for i in range(10):
    t = threading.Thread(target=mandar_imprimr, args=(i,))
    t.start()
    hilos.append(t)


for t in hilos:
    t.join()

print('Operaciones terminadas')
```

**Esto nos da a entender que todas las solicitudes se procesaron simultaneamente, lo cual seria imposible con solo 2 impresoras.**
