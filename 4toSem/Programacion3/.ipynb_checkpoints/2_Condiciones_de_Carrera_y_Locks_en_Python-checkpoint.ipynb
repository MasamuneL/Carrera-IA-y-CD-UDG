# Librerías nuevas para la clase

Usaremos `threading.Lock`, que forma parte de la misma
librería estándar `threading` que vimos antes, pero con un propósito distinto.

1. threading.Lock()
   Un Lock es un mecanismo de sincronización muy simple pero muy útil.
   Sirve para evitar que dos o más hilos ejecuten una sección crítica al mismo tiempo.
   Conceptualmente funciona como un candado:
   - Solo un hilo puede "adquirirlo" a la vez.
   - Los demás hilos deben esperar.

 ### Uso:
       lock = threading.Lock()

   ### Para usar el candado dentro de una sección crítica:
       with lock:
           # código que solo un hilo puede ejecutar a la vez

# Recordatorio

Los hilos pueden ejecutar tareas en paralelo lógico (concurrencia).
Pero cuando comparten datos entre sí, pueden surgir problemas.

Hoy vamos a ver un ejemplo clásico de error en concurrencia:
**la condición de carrera**.

Una condición de carrera ocurre cuando:
- Dos o más hilos acceden al mismo recurso
- Al menos uno lo modifica
- El resultado depende del "orden" en que se ejecuten los hilos,
  un orden que no podemos controlar

## Veamos un ejemplo clásico de condición de carrera.

 Vamos a crear un contador compartido.
 Luego, varios hilos lo incrementarán muchas veces.
 Veremos que el resultado final NO es el esperado.

```python
# Ejecutemos varias veces este código para ver lo que pasa (o también podemos hacer un ciclo con varias ejecuciones)

import threading
import time

contador = 0

def incrementar():
    global contador
    for _ in range(100000):
        local_contador = contador
        time.sleep(0.0001)
        contador = local_contador + 1

# Creamos varios hilos
hilos = []
for _ in range(5):
    t = threading.Thread(target=incrementar)
    t.start()
    hilos.append(t)

# Esperamos a que terminen
for t in hilos:
    t.join()

print("Valor esperado:", 5 * 100000)
print("Valor real:", contador)
```

## ¿Qué ocurrió?

La operación contador += 1 NO es atómica.
### Significa que internamente ocurre en tres pasos:
   1. Leer contador
   2. Sumar 1
   3. Guardar el resultado

### Si varios hilos se "interrumpen" mutuamente entre estos pasos, algunos incrementos se pierden.

 Esto es una condición de carrera: el resultado depende del orden en que los hilos se intercalan, y no podemos controlar ese orden.

### Esta es la definición práctica de una **condición de carrera**.

## Vamos a corregir el problema usando algo llamdo Lock (candado).
Un Lock asegura que sólo un hilo a la vez pueda ejecutar una sección crítica.



```python
import threading
import time

contador = 0
lock = threading.Lock()

def incrementar_seguro():
    global contador
    for _ in range(100000):
        # Este 'with lock' garantiza acceso exclusivo a la variable 'contador'
        with lock:
          time.sleep(0.0001)
          contador += 1

hilos = []
for _ in range(5):
    t = threading.Thread(target=incrementar_seguro)
    t.start()
    hilos.append(t)

for t in hilos:
    t.join()

print("Valor esperado:", 5 * 100000)
print("Valor real (con lock):", contador)
```

Podemos ver que no importa la cantidad de veces que ejecutemos el código, el resultado siempre es consistente.

## Ejemplo práctico:
### Simulemos una cuenta bancaria donde varios hilos retiran dinero.

```python
import threading
import time

saldo = 100

def retirar(monto):
    global saldo
    actual = saldo
    time.sleep(0.000001)  # simulamos una operación que tarda
    saldo = actual - monto

hilos = []
for _ in range(10):
    t = threading.Thread(target=retirar, args=(10,))
    t.start()
    hilos.append(t)

for t in hilos:
    t.join()

print("Saldo esperado: 0\n")
print("Saldo final:", saldo)
```

En este caso, el saldo será incorrecto. A veces tendremos saldo negativo, otras saldo arriba de cero, pero siempre el resultado es inconsistente.

### Banco con Lock
Volvamos a realizar el ejercicio, pero ahora introduciendo un candado que permita que no se translapen las operaciones de los hilos

```python
saldo = 100
lock = threading.Lock()

def retirar_seguro(monto):
    global saldo
    with lock:
        actual = saldo
        time.sleep(0.001)
        saldo = actual - monto

hilos = []
for _ in range(10):
    t = threading.Thread(target=retirar_seguro, args=(10,))
    t.start()
    hilos.append(t)

for t in hilos:
    t.join()

print("Saldo final (correcto):", saldo)
```

# Por lo tanto:

- Cuando varios hilos modifican el mismo recurso, pueden surgir errores difíciles de detectar.
 - Estos errores se llaman condiciones de carrera.
 - Un Lock permite que solo un hilo acceda a una sección crítica a la vez.
 - Usar locks es una herramienta fundamental para escribir programas concurrentes correctos.
