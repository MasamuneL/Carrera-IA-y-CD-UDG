```python
#Importamos Librerias
import threading
import time
import random
```

```python
#Creamos la clase Inventario

class inventario:
    
    def __init__(self, stock):
        self.local_stock = stock
        self.lock = threading.Lock()

    def stock_add(self, a):
        """
        Agrega 'a' cantidad de cosas al stock
        """
        with self.lock:
            time.sleep(0.0001)
            self.local_stock = self.local_stock + a
            
    def stock_sub(self, b):
        """
        Substrae 'b' cantidad de cosas al stock
        """
        with self.lock:
            time.sleep(0.0001)
            self.local_stock = self.local_stock - b
```

```python
# Creamos el Inventario
chocolates = inventario(50)

# Creamos varios hilos
hilos = []
for _ in range(5):
    t = threading.Thread(target=chocolates.stock_add, args=(random.randrange(2,10),))
    t.start()
    hilos.append(t)

# Esperamos a que terminen
for t in hilos:
    t.join()

print(f'El stock de chocolates actual es de : {chocolates.local_stock} chocolates')
```

```python
# Hacemos lo mismo pero ahora con la resta

hilos = []
for _ in range(5):
    t = threading.Thread(target=chocolates.stock_sub, args=(random.randrange(2,10),))
    t.start()
    hilos.append(t)

# Esperamos a que terminen
for t in hilos:
    t.join()

print(f'El stock de chocolates final es de : {chocolates.local_stock} chocolates')
```
