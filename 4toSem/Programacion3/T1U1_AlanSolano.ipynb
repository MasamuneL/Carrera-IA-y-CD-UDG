#### Ejercicio 1 (Sequencial)

```python
# Contador de palabras sequencial

import requests
import time

def contar_palabras(url):
    print(f"Contando palabras de: {url}")
    try:
        r = requests.get(url)
        r.encoding = 'utf-8' # Asegura que lea acentos correctamente
        texto = r.text # Obtenemos el texto de la página
        
        counts = dict()
        palabras = texto.split() # Dividimos todo el texto en palabras
        
        for palabra in palabras:
            counts[palabra] = counts.get(palabra, 0) + 1
            
        print(f"Completado: {url} | Total palabras únicas: {len(counts)}")
        # Nota: Imprimir el diccionario completo 'counts' puede ser enorme en webs grandes.
        
    except Exception as e:
        print(f"Error al conectar con {url}: {e}")

urls = ["https://siiau.udg.mx", "https://www.youtube.com/"]

print("Contando sin hilos")
inicio = time.time()

for u in urls:
    contar_palabras(u)

fin = time.time()
print(f"Tiempo total: {fin - inicio:.4f} segundos")
```

#### Ejercicio 2 (Concurrente)

```python
# Actividad: Descarga concurrente con hilos

import threading
import requests
import time

def contar_palabras(url):
    print(f"Contando palabras de: {url}")
    try:
        r = requests.get(url)
        r.encoding = 'utf-8' # Asegura que lea acentos correctamente
        texto = r.text # Obtenemos el texto de la página
        
        counts = dict()
        palabras = texto.split() # Dividimos todo el texto en palabras
        
        for palabra in palabras:
            counts[palabra] = counts.get(palabra, 0) + 1
            
        print(f"Completado: {url} | Total palabras únicas: {len(counts)}")
        # Nota: Imprimir el diccionario completo 'counts' puede ser enorme en webs grandes.
        
    except Exception as e:
        print(f"Error al conectar con {url}: {e}")
        
urls = ["https://siiau.udg.mx","https://www.youtube.com/"]

hilos = []

print("Contando con hilos")
inicio = time.time()

for url in urls:
    t = threading.Thread(target=contar_palabras, args=(url,))
    t.start()
    hilos.append(t)

for t in hilos:
    t.join()
    
fin = time.time()
print(f"Tiempo total: {fin - inicio:.4f} segundos")
```
