### Suma de rangos con hilos
Alan Josue Solano Perez

```python
import threading
import time

def rangos(n, h):
    resultados = []
    m = n//h
    for i in range(h):
        inicio = i*m
        fin = n if i == h -1 else inicio + m
    suma = sum(range(inicio,fin))
    resultados.append(suma)
    return resultados


    

n = input('Ingrese el numero hasta el que desea contar: ')
h = input("Ingrese el numero de hilos: ")
try:
        n = int(n)
        h = int(h)
except ValueError:
        print("Ingrese un valor valido (numeros enteros positivos)")
    
hilos = []

for i in range(h):
    t = threading.Thread(target=rangos, args=(n,h))
    hilos.append(t)
    t.start()

for t in hilos:
    t.join()
```
