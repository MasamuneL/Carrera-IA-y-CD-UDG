# Introducción a los métodos ARIMA para series de tiempo

Los modelos **ARIMA** (AutoRegressive Integrated Moving Average) constituyen una de las herramientas más utilizadas para modelar y pronosticar series de tiempo estacionarias o que pueden transformarse en estacionarias mediante diferenciación. Su fortaleza radica en combinar tres componentes fundamentales: autorregresión, diferenciación e información de medias móviles.

Un modelo ARIMA se denota como  
\begin{equation}
\text{ARIMA}(p, d, q),
\end{equation}
donde:

- **\(p\)**: orden del componente *autorregresivo* (AR)  
- **\(d\)**: número de diferencias aplicadas a la serie para hacerla estacionaria  
- **\(q\)**: orden del componente de *media móvil* (MA)

---

## Componente AR (AutoRegresivo)

El modelo AR de orden \(p\) expresa que el valor actual depende de sus propios rezagos:
\begin{equation}
y_t = \phi_1 y_{t-1} + \phi_2 y_{t-2} + \dots + \phi_p y_{t-p} + \varepsilon_t,
\end{equation}
donde $\varepsilon_t$ es un ruido blanco.

## Componente I (Integrado)

La parte **I** consiste en diferenciar la serie. La primera diferencia es:
\begin{equation}
\nabla y_t = y_t - y_{t-1}.
\end{equation}
Aplicar $d$ diferencias intenta eliminar tendencias y volver la serie estacionaria.

## Componente MA (Media Móvil)

Un modelo MA de orden \(q\) utiliza los rezagos del error:
\begin{equation}
y_t = \mu + \varepsilon_t + \theta_1 \varepsilon_{t-1} + \theta_2 \varepsilon_{t-2} + \dots + \theta_q \varepsilon_{t-q}.
\end{equation}

---

### Cómo se ve un modelo ARIMA completo

Un ARIMA(p, d, q) combina los tres elementos anteriores de la siguiente forma:

\begin{equation}
\nabla^d y_t = 
\alpha 
+ \sum_{i=1}^{p} \phi_i \nabla^d y_{t-i}
+ \sum_{j=1}^{q} \theta_j \varepsilon_{t-j}
+ \varepsilon_t.
\end{equation}

Tu pronóstico será entonces la suma de:

- un intercepto,
- los valores previos de la serie,
- los errores previos del modelo,
- y un término aleatorio.

---

```python
# Importación de librerías
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# statsmodels
from statsmodels.graphics.tsaplots import month_plot, quarter_plot, plot_acf, plot_pacf
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.statespace.sarimax import SARIMAX

# scikit-learn
from sklearn.metrics import mean_absolute_error, root_mean_squared_error, mean_absolute_percentage_error
from sklearn.model_selection import ParameterGrid, TimeSeriesSplit
```

```python
## Funciones útiles

def model_plot(train, test, predictions, chart_title):
    
    plt.figure(figsize=(10,4))
    plt.plot(train, label='Train')
    plt.plot(test, label='Test')
    plt.plot(predictions, label='Forecast')
    plt.title(f'Entrenamiento, Validación, y Predicciones con {chart_title}')
    plt.legend()
    plt.show()

    mae = mean_absolute_error(test, predictions)
    rmse = root_mean_squared_error(test, predictions)
    mape = mean_absolute_percentage_error(test, predictions)

    print(f"MAE: {mae:.2f}")
    print(f"RMSE: {rmse:.2f}")
    print(f"MAPE: {100*mape:.2f} %")
```

```python
def plot_future(y, forecast, title):
    
    plt.figure(figsize=(10,4))
    plt.plot(y, label='Train')
    plt.plot(forecast, label='Forecast')
    plt.title(f'Entrenamiento y Pronóstico con {title}')
    plt.legend()
    plt.show()
```

## Prueba de hipótesis (Dickey-Fuller aumentada, ADF) para proceso estacionario


La **prueba de Dickey–Fuller aumentada (ADF)** es una herramienta para determinar si una serie de tiempo es **estacionaria**. En el contexto de modelos ARIMA, esta prueba es especialmente importante porque la estacionariedad es un requisito para que el modelo capture de forma adecuada la dinámica de la serie.

La prueba ADF evalúa si una serie posee una **raíz unitaria**, lo que implica no estacionariedad. Para ello, analiza si el valor actual de la serie puede expresarse en función de su valor rezagado y de sus diferencias.

---

### Formulación de la prueba

La forma ampliada del modelo evaluado en la prueba es:

\begin{equation}
\Delta y_t = \alpha + \beta t + \gamma y_{t-1}
+ \sum_{i=1}^{k} \delta_i \Delta y_{t-i} + \varepsilon_t,
\end{equation}

donde:

- $\Delta y_t = y_t - y_{t-1}$ es la primera diferencia,  
- $t$ es una tendencia opcional,  
- $k$ es el número de rezagos incluidos para controlar autocorrelación,  
- $\gamma$ es el parámetro clave para determinar estacionariedad.

---

### Hipótesis de la prueba

La prueba ADF contrasta:

- **Hipótesis nula $H_0$**: la serie tiene raíz unitaria  
  \begin{equation}
  H_0: \gamma = 0 \quad \Rightarrow \quad \text{la serie no es estacionaria}
  \end{equation}

- **Hipótesis alternativa $H_1$**: la serie es estacionaria  
  \begin{equation}
  H_1: \gamma < 0 \quad \Rightarrow \quad \text{la serie es estacionaria}
  \end{equation}

---

### Interpretación del resultado

La librería `statsmodels` devuelve:

- el **estadístico ADF**,  
- el **p-value**,  
- los valores críticos al 1%, 5% y 10%,  
- el número óptimo de rezagos usados.

Para interpretar:

1. **Si el p-value < nivel de significancia (por ejemplo, 0.05)**  
   Se rechaza $H_0$: la serie **es estacionaria**.

2. **Si el p-value ≥ nivel de significancia**  
   No se puede rechazar $H_0$: la serie **no es estacionaria**.

También puede compararse el estadístico ADF con los valores críticos.

---

### Uso en el flujo ARIMA

La prueba ADF te ayudará a decidir el valor adecuado de $d$ (orden de diferenciación):

- Si la serie no es estacionaria, se aplican diferencias.  
- Tras diferenciación, se vuelve a aplicar la prueba ADF para verificar.  

Este proceso continúa hasta obtener estacionariedad sin sobre-diferenciar la serie.

```python
# Prueba de adfuller
adfuller(df.y)
```

```python
result = adfuller(df.y)
print(f"p-value: {result[1]}")

significance = 0.05
if result[1] > significance:
    print("La serie de tiempo no es estacionaria")
else:
    print("Es probable que la serie de tiempo sea estacionaria")
```

```python
# Diferenciación para df.y
df['y_diff'] = df['y'].diff()
df.head()
```

```python
# Prueba después de la diferenciación

result = adfuller(df.y_diff.dropna()) # le quitamos los primero valores nulos debido al método
print(f"p-value: {result[1]}")

significance = 0.05
if result[1] > significance:
    print("La serie de tiempo no es estacionaria")
else:
    print("Es probable que la serie de tiempo sea estacionaria")
```

```python
# Plot simple
df['y_diff'].plot(title='Ganancia diaria (diff)')
plt.show()
```

## Intuición del componente de Media Móvil (MA)

Para completar la construcción del modelo ARIMA, en esta sección estudiarás el componente **MA (Moving Average)**. Esta parte del modelo introduce una idea clave: **aprender de los errores pasados**. Para entenderla, piensa en el siguiente ejemplo.

Imagina que quieres estimar cuánta **cafeína** o **tazas de café** vas a necesitar mañana.  Una forma básica de hacerlo sería revisar cuánto café consumiste ayer y días previos. Esa es la lógica autorregresiva: usar valores pasados de la serie.

Sin embargo, existe otra fuente de información útil: **las equivocaciones que cometiste al estimar tu consumo en el pasado**.

Por ejemplo:

- Quizá ayer pensaste que ibas a tomar tres tazas,  
- pero en realidad solo tomaste una.

Esa diferencia entre lo que esperabas y lo que ocurrió —el **error de predicción**— contiene información valiosa. El componente MA integra justamente estos errores recientes para mejorar los pronósticos futuros.

---

### ¿Qué hace realmente el componente MA?

En términos simples:

- El modelo **MA observa los errores pasados del modelo**,  
- y los usa para ajustar la predicción actual.

La forma general de un **MA(q)** es:

\begin{equation}
y_t = \mu + \varepsilon_t + \theta_1 \varepsilon_{t-1} +
\theta_2 \varepsilon_{t-2} + \dots + \theta_q \varepsilon_{t-q},
\end{equation}

donde:

- $\varepsilon_t$ es el error aleatorio actual,  
- $\varepsilon_{t-i}$ son errores de pasos anteriores,  
- $\theta_i$ controla cuánto pesa cada error pasado.

---

### ¿Por qué es útil?

El componente MA permite:

1. **Suavizar fluctuaciones aleatorias.**  
   Cuando la serie presenta “saltos” o ruido, los errores ayudan a corregir rápidamente las desviaciones recientes.

2. **Ajustes rápidos ante cambios repentinos.**  
   Si hay una caída o subida inesperada, los errores pasados capturan esa sorpresa y el modelo se adapta sin depender únicamente de valores previos.

3. **Mejorar la precisión del pronóstico.**  
   No solo se aprende de lo que ocurrió (valores pasados), sino también de las **equivocaciones recientes**.

En otras palabras, el MA añade una capa de “sabiduría”: recuerda no solo los datos, sino los errores cometidos al modelarlos. Esto permite predicciones más estables y responsivas.

---

Con esta intuición del componente MA ya cuentas con las tres piezas esenciales de ARIMA:  
- Autorregresión (AR),  
- Integración (I),  
- Media Móvil (MA),  

Cada una aporta información distinta y complementaria para construir modelos de series de tiempo más robustos.

--

## Introducción a los modelos SARIMA

Hasta ahora has visto que los modelos ARIMA permiten capturar dependencia temporal mediante tres componentes: autorregresión (AR), diferenciación (I) y medias móviles (MA). Sin embargo, muchos fenómenos reales presentan **estacionalidad**, es decir, patrones que se repiten regularmente con una frecuencia fija: meses, semanas, trimestres, estaciones del año, etc.

Para este tipo de series, ARIMA no es suficiente. Aquí entra en juego **SARIMA**: *Seasonal ARIMA*. Un modelo **SARIMA** extiende ARIMA incorporando un segundo conjunto de términos AR, I y MA, pero aplicados sobre la estructura estacional de la serie.

---

### Notación general de SARIMA

Un modelo SARIMA se escribe como:

\begin{equation}
\text{SARIMA}(p,d,q)\times(P,D,Q)_s,
\end{equation}

donde:

- $p, d, q$: órdenes del modelo **no estacional**, como en ARIMA,
- $P, D, Q$: órdenes del **componente estacional**,
- $s$: longitud del periodo estacional (por ejemplo, $s=12$ para datos mensuales con estacionalidad anual).

---

### Componentes estacionales del modelo

SARIMA agrega un conjunto adicional de transformaciones:

1. **Diferencias estacionales (D)**  
   Para series con un patrón que se repite cada \(s\) pasos, se usa:
   \begin{equation}
   \nabla_s y_t = y_t - y_{t-s}.
   \end{equation}
   Esto elimina patrones repetitivos y permite estabilizar la serie.

2. **Autorregresión estacional (P)**  
   Similar al AR tradicional, pero utilizando valores rezagados múltiples de \(s\):
   \begin{equation}
   y_t = \Phi_1 y_{t-s} + \Phi_2 y_{t-2s} + \dots + \varepsilon_t.
   \end{equation}

3. **Medias móviles estacionales (Q)**  
   Igual que los errores lag en MA, pero ajustados por \(s\):
   \begin{equation}
   y_t = \varepsilon_t + \Theta_1 \varepsilon_{t-s}
   + \Theta_2 \varepsilon_{t-2s} + \dots.
   \end{equation}

Estos términos permiten capturar patrones recurrentes, como ventas mensuales, consumo energético semanal, visitas diarias en sitios web, entre otros.

---

### Ecuación conceptual del modelo completo

Un SARIMA combina ambas estructuras:

- ARIMA para la dinámica “inmediata” o de corto plazo,
- ARIMA estacional para los patrones de largo plazo con periodicidad fija.

De manera compacta, el modelo se puede expresar en notación operatoria como:

\begin{equation}
\Phi(B^s)\, \phi(B)\, (1-B)^d (1-B^s)^D \, y_t
=
\Theta(B^s)\, \theta(B)\, \varepsilon_t,
\end{equation}

donde:

- $B$ es el operador de rezago,
- $\phi(B)$ y $\theta(B)$: polinomios AR y MA no estacionales,
- $\Phi(B^s)$ y $\Theta(B^s)$: polinomios estacionales.

Esta formulación muestra cómo se combinan ambas estructuras en un mismo modelo.

---

### ¿Cuándo usar SARIMA?

Utiliza SARIMA cuando:

- La serie muestra **estacionalidad clara** y recurrente.  
- La estacionalidad no puede eliminarse solo con diferencias no estacionales.  
- Necesitas un modelo capaz de capturar variaciones locales y globales (corto y largo plazo).

SARIMA te permitirá generar pronósticos que respeten las fluctuaciones periódicas de la serie, manteniendo al mismo tiempo la estructura ARIMA tradicional.

---

En las siguientes secciones aplicarás SARIMA con `statsmodels`.

## Introducción a los modelos SARIMAX

Los modelos **SARIMAX** (*Seasonal ARIMA with eXogenous regressors*) amplían la capacidad de SARIMA al permitir incorporar **variables externas** (también llamadas regresores exógenos) que influyen en la serie. Esto resulta especialmente útil cuando el comportamiento del fenómeno no depende únicamente de su propio pasado, sino también de otros factores observables.

Por ejemplo:

- ventas afectadas por campañas publicitarias,  
- consumo energético influido por la temperatura,  
- tráfico o demanda afectada por eventos especiales,  
- series macroeconómicas que dependen de indicadores externos.

Con SARIMAX puedes integrar esta información para mejorar la precisión del pronóstico.

---

### Notación del modelo

Un modelo SARIMAX se expresa como:

\begin{equation}
\text{SARIMAX}(p,d,q)\times(P,D,Q)_s + \text{X},
\end{equation}

donde los primeros términos corresponden al modelo SARIMA y **X** representa el conjunto de regresores exógenos.

---

### Expresión general del modelo

En forma compacta:

\begin{equation}
y_t =
\mu +
\mathbf{x}_t^\top \boldsymbol{\beta}
+
\text{(componentes ARIMA no estacionales)}
+
\text{(componentes ARIMA estacionales)}
+
\varepsilon_t.
\end{equation}

Aquí:

- $\mathbf{x}_t$ es el vector de variables exógenas en el tiempo \(t\),  
- $\boldsymbol{\beta}$ son sus coeficientes,  
- $\varepsilon_t$ es el término de error.

El resto de la estructura corresponde a los componentes ARIMA y SARIMA que ya se describieron anteriormente.

---

### ¿Para qué sirven los regresores exógenos?

Las variables externas permiten capturar efectos que el modelo univariado no podría explicar por sí solo. Algunos usos típicos:

1. **Agregar información adicional** que mejora el pronóstico.  
2. **Controlar variables de confusión** que influyen en la serie.  
3. **Modelar impactos puntuales** (por ejemplo, un feriado o un shock).  
4. **Incluir variables futuras conocidas** (por ejemplo, calendario, clima pronosticado).

El modelo aprende cómo estos factores afectan la serie y los incluye explícitamente en el pronóstico.

---

### Importancia del alineado temporal

Es fundamental que las variables exógenas:

- estén alineadas en el tiempo con la serie objetivo,  
- no contengan información futura (a menos que sea conocida con anticipación),  
- estén escaladas o transformadas según corresponda.

---

### Cuándo usar SARIMAX

Usa SARIMAX cuando:

- La serie presenta **estacionalidad**,  
- Existe dependencia temporal que requiere ARIMA/SARIMA,  
- Hay **variables externas relevantes** para explicar o mejorar las predicciones.

En contextos reales, SARIMAX suele superar a SARIMA cuando dispones de información complementaria confiable.

## Estacionalidad débil en SARIMA y SARIMAX

### Qué significa estacionalidad “débil”

En SARIMA (Seasonal ARIMA) o SARIMAX (que agrega regresores exógenos), la estacionalidad se modela mediante **diferenciación estacional** y **términos autorregresivos y de media móvil estacionales**, que operan con un **periodo fijo $s$**. Por ejemplo, si $s = 12$ para datos mensuales, el modelo considera relaciones entre $y_t$ y $y_{t-12}$.

> Sin embargo, esto **no significa** que el modelo "recuerde" los valores de hace un año completo, sino que **ajusta una relación lineal** con el pasado inmediato en ese rezago estacional.

Por eso se llama *estacionalidad débil*:  
- La estacionalidad se refleja en correlaciones estadísticas de corto plazo, no en ciclos determinísticos.  
- Si los patrones anuales cambian lentamente (como ventas que aumentan año con año), el modelo **no los aprenderá bien**, porque asume que la estructura de la serie es *estacionaria después de la diferenciación*.

SARIMA y SARIMAX solo consideran un número **finito y fijo de rezagos** (por ejemplo, $p$, $q$, $P$, $Q$ y $s$).  
Esto significa que el modelo no “recuerda” los valores de hace más de $s$ pasos atrás, salvo lo que está implícitamente codificado en la estructura lineal.

- Si los patrones anuales **no son perfectamente repetitivos**, el modelo pierde la capacidad de proyectarlos.  
- No puede capturar **tendencias a largo plazo** o **cambios en la forma de la estacionalidad** (por ejemplo, que la amplitud del ciclo crezca o disminuya cada año).  
- La predicción “más allá de un ciclo” (por ejemplo, a 24 o 36 pasos) se construye *recursivamente*, por lo que el error acumulado se amplifica.

### Ejemplo

Imagina una serie de ventas con picos cada diciembre, pero con un crecimiento constante año con año:

| Año | Diciembre |
|------|------------|
| 2020 | 200 |
| 2021 | 250 |
| 2022 | 300 |

SARIMA con $s=12$ solo aprende “hay un pico en diciembre”, no “el pico aumenta 50 cada año”.  
Entonces, su predicción para 2023 tenderá a repetir el patrón de 2022, sin incorporar la tendencia de crecimiento.

## Validación cruzada en series de tiempo (*Sliding* vs *Rolling*)

En series de tiempo, no se puede usar la validación cruzada tradicional porque los datos están ordenados temporalmente y el futuro **no debe influir en el pasado**.  Por eso se usan esquemas especiales: **sliding** y **rolling cross-validation**.

---

### *Rolling Cross-Validation* (ventana expansiva)

En este método, el conjunto de entrenamiento **se expande** con cada iteración, incorporando más observaciones pasadas.  
El modelo se entrena con todos los datos disponibles hasta un punto y se evalúa en el siguiente bloque temporal.

Ejemplo:

| Iteración | Entrenamiento              | Prueba     |
|------------|---------------------------|-------------|
| 1          | [1, 2, 3]                 | [4]         |
| 2          | [1, 2, 3, 4]              | [5]         |
| 3          | [1, 2, 3, 4, 5]           | [6]         |


**Ventajas:**  
- Usa más información conforme avanza el tiempo.  
- Ideal para modelos donde los patrones son estables o acumulativos.

**Desventajas:**  
- Puede ser costoso computacionalmente.  
- Menos adecuado si la serie presenta cambios estructurales fuertes.

---

### *Sliding Cross-Validation* (ventana deslizante)

Aquí la ventana de entrenamiento **mantiene un tamaño fijo**.  
A medida que avanza el tiempo, se descartan los datos más antiguos y se agregan los más recientes.

Ejemplo:

| Iteración | Entrenamiento        | Prueba     |
|------------|---------------------|-------------|
| 1          | [1, 2, 3]           | [4]         |
| 2          | [2, 3, 4]           | [5]         |
| 3          | [3, 4, 5]           | [6]         |


**Ventajas:**  
- Refleja mejor entornos donde los patrones cambian con el tiempo.  
- Menor carga de cómputo.

**Desventajas:**  
- Pierde información histórica.  
- Puede ser menos estable si el tamaño de ventana es pequeño.

---


```python
# Configuración de la validación cruzada
order = (3,1,1)
seasonal_order = (2,0,1,7)
n_splits = 5
```

```python
tscv = TimeSeriesSplit(n_splits=n_splits,
                       test_size=test_days)
```

```python
# Inspeccionamos los splits
for train_index, test_index in tscv.split(df):
    print(f"Train: {train_index}, Test: {test_index}")
```

```python
# Implementacion del split para CV, basado en RMSE, MAE, MAPE

cv_scores = []

for train_index, test_index in tscv.split(df):
    train = df.iloc[train_index]
    test = df.iloc[test_index]

    # Data Exog
    train_reg = df[["discount_rate", "coupon_rate"]].iloc[train_index]
    test_reg = df[["discount_rate", "coupon_rate"]].iloc[test_index]
    
    # SARIMAX
    model_sarimax = SARIMAX(
        endog=train['y'],
        exog=train_reg,
        order=order,
        seasonal_order=seasonal_order
    ).fit()

    # Predicciones
    predictions_sarimax = model_sarimax.forecast(steps=test_days,
                                                 exog=test_reg)

    # Evaluacion, metricas
    rmse = root_mean_squared_error(test['y'], predictions_sarimax)
    mae = mean_absolute_error(test['y'], predictions_sarimax)
    mape = mean_absolute_percentage_error(test['y'], predictions_sarimax)

    # Guardar en lista
    cv_scores.append(
        {
            'rmse':rmse,
            'mae':mae,
            'mape':mape
        }
    )
```

```python
# Mostramos los resultados
cv_scores = pd.DataFrame(cv_scores)
cv_scores.index.name = 'Fold'
print(cv_scores)
```

```python
# Mostramos los errores promedios
print(f"RMSE promedio: {cv_scores['rmse'].mean():.2f}")
print(f"MAE promedio: {cv_scores['mae'].mean():.2f}")
print(f"MAPE promedio: {cv_scores['mape'].mean():.2f}")
```

## Ajuste de parámetros

```python
param_grid = {
    'p':[1,3],
    'd':[1],
    'q':[0,1],
    'P':[1,2],
    'D':[0],
    'Q':[1]
}
grid = ParameterGrid(param_grid=param_grid)
print(f"número de combinaciones: {len(list(grid))}")
list(grid)
```

```python
# split de la serie de tiempo
n_splits = 5
tscv = TimeSeriesSplit(n_splits=n_splits,
                       test_size=test_days)
```

```python
# Implementación de los modelos con el grid de parametros
rmse_list = []

# Iteramos sobre el grid de parámetros
for params in grid:
  fold_rmse = []

  # Construimos el modelo
  for train_index, test_index in tscv.split(df):
    train = df.iloc[train_index]
    test = df.iloc[test_index]

    # Datos Exog
    train_reg = df[["discount_rate", "coupon_rate"]].iloc[train_index]
    test_reg = df[["discount_rate", "coupon_rate"]].iloc[test_index]

    # SARIMAX
    model_sarimax = SARIMAX(
        train['y'],
        exog = train_reg,
        order = (params['p'], params['d'], params['q']),
        seasonal_order = (params['P'], params['D'], params['Q'], 7)).fit()

    # Calculamos el error y guardamos las predicciones
    predictions_sarimax = model_sarimax.forecast(steps = test_days,
                                                  exog = test_reg)
    rmse = root_mean_squared_error(test['y'], predictions_sarimax)
    fold_rmse.append(rmse)

  # Guardamos los resultados
  rmse_list.append(np.mean(fold_rmse))
```

```python
# Checamos los resultados, agregamos el resultado al diccionario del grid y transformamos todo a un dataframe
rmse_df = pd.DataFrame(list(grid))
rmse_df['rmse'] = rmse_list
rmse_df
```

```python
# Extraemos los mejores parametros
best_params = rmse_df.loc[rmse_df['rmse'].idxmin()].to_frame()
best_params

# Podemos exportar estos parametros
#best_params.to_csv("best_params_sarimax.csv")
```

## Pronóstico para datos futuros

```python
# Utilizamos todos los datos para predecir
# Separamos en variable dependiente e independientes
y = df['y']
X = df[["discount_rate", "coupon_rate"]]
```

```python
# Utilizamos los mejores parametros que encontramos
p = int(best_params.loc['p'])
d = int(best_params.loc['d'])
q = int(best_params.loc['q'])
P = int(best_params.loc['P'])
D = int(best_params.loc['D'])
Q = int(best_params.loc['Q'])
```

```python
# Cargamos las variables independientes futuras (tienen que continuar los indices historicos del dataset de entrenamiento)
future_reg = pd.read_csv("data/future_regressors.csv", index_col = 0, dayfirst = True, parse_dates = True)
future_reg.head()
```

```python
# Separamos las variables independientes, en este caso discount_rate y coupon_rate, de estos datos futuros
# Mantenemos el mismo escalado que habiamos utilizado, como porcentajes
X_future = future_reg[["discount_rate", "coupon_rate"]] * 100
X_future.head()
```

```python
# Construimos el modelo SARIMAX
model_sarimax = SARIMAX(
    y,
    exog = X,
    order = (p,d,q),
    seasonal_order = (P,D,Q,7)).fit()
print(model_sarimax.summary())
```

```python
# predicciones a futuro
future_predictions = model_sarimax.forecast(steps = len(X_future),
                                             exog = X_future)
```

```python
# Graficamos las predicciones y comparamos
plot_future(y['2022'], future_predictions, "SARIMAX")
```

Una recomendación es que si queremos incluir efectos de fechas donde hay aumento de ventas, utilizamos algún indicador binario. Además, recordamos que nuestra ventana de predicción tiene que ser consistente con el uso que se la va a dar. Por ejemplo, en este caso siempre estamos haciendo evaluaciones de predicciones a 30 días, entonces mantenemos esta misma ventana para predicciones a futuro.
