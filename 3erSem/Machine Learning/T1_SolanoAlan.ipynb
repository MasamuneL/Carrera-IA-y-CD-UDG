### Acerca del conjunto de datos

Segun el autor la base datos contiene lo siguiente:

**Información general**

Esta es una colección de más de 50 000 partidas clasificatorias de la EUW de League of Legends, así como archivos JSON que contienen una forma de convertir los identificadores y nombres de los hechizos de campeón e invocador. Cada partida incluye campos para:

- ID de la partida
- Hora de creación (en formato de época)
- Duración de la partida (en segundos)
- ID de la temporada
- Ganador (1 = equipo 1, 2 = equipo 2)
- Primer Barón, Dragón, Torre, Sangre, Inhibidor y Heraldo de la Grieta (1 = equipo 1, 2 = equipo 2, 0 = ninguno)
- Campeones y hechizos de invocador de cada equipo (almacenados como los ID de campeón y hechizo de invocador de Riot)
- Número de bajas de cada equipo por torre, inhibidor, Barón, Dragón y Heraldo de la Grieta
- 5 baneos de cada equipo (de nuevo, se utilizan los ID de campeón)

Este conjunto de datos se recopiló mediante la API de Riot Games, lo que facilita la búsqueda y recopilación de información sobre el historial de clasificatorias de los usuarios y sus partidas. Sin embargo, encontrar una lista de nombres de usuario es la parte difícil; en este caso, utilizo una lista de nombres de usuario extraída de sitios web de LoL de terceros.

```python
#separacion de variables mediante tipo cuantitativo o cualitativo
num_vars = df_games.select_dtypes(include=[np.number]).columns
cat_vars = df_games.select_dtypes(exclude=[np.number]).columns

print(f"Variable cuantitativas: {list(num_vars)}")
print(f"Variable cualitativas: {list(cat_vars)}")
```
