### **1. Gráfica de la Distribución F**
**Pregunta 1:** Grafique (en python) la distribución F para los grados de libertad $(\nu_{1},\nu_{2})=(1,1),$ $(\nu_{1},\nu_{2})=(2,1)$, $(\nu_{1},\nu_{2})=(5,2)$.

```python
# Importar librerías necesarias
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f, chi2

# Configuración para que las gráficas se vean mejor
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
# Definir el rango de valores para el eje x
# La distribución F solo está definida para valores > 0
x = np.linspace(0.01, 5, 500)

# Pares de grados de libertad a graficar
grados_libertad = [(1, 1), (2, 1), (5, 2)]

# Crear la figura para la gráfica
plt.figure(figsize=(12, 7))

# Graficar las distribuciones F para cada par de grados de libertad
for df1, df2 in grados_libertad:
    pdf_f = f.pdf(x, dfn=df1, dfd=df2)
    plt.plot(x, pdf_f, label=f'F (ν₁={df1}, ν₂={df2})')

# Añadir títulos y etiquetas
plt.title('Función de Densidad de Probabilidad de la Distribución F')
plt.xlabel('Valor F')
plt.ylabel('Densidad de Probabilidad')
plt.ylim(0, 1.0) # Ajustar el límite del eje y para mejor visualización
plt.legend()
plt.show()
```

**Pregunta 2:** La afirmación de que la varianza de una población normal es $\sigma^{2}=25$ debe rechazarse si la varianza de una muestra aleatoria de tamaño 16 excede 54.668 o es menor a 12.102. ¿Cuál es la probabilidad de que esta afirmación se rechazará aún cuando $\sigma^{2}=4$? 

```python
# Datos del problema
n = 16
df = n - 1
true_variance = 4
s2_lower = 12.102
s2_upper = 54.668

# Convertir los límites de s^2 a valores de Chi-cuadrado
chi2_lower_val = (df * s2_lower) / true_variance
chi2_upper_val = (df * s2_upper) / true_variance

# Calcular las probabilidades
prob_lower = chi2.cdf(chi2_lower_val, df=df)
prob_upper = 1 - chi2.cdf(chi2_upper_val, df=df) # o chi2.sf(chi2_upper_val, df=df)

total_prob = prob_lower + prob_upper

print(f"Grados de libertad (ν): {df}")
print(f"Límite Chi-cuadrado inferior: {chi2_lower_val:.4f}")
print(f"Límite Chi-cuadrado superior: {chi2_upper_val:.4f}")
print("-" * 30)
print(f"P(Y < {chi2_lower_val:.4f}) = {prob_lower:.6f}")
print(f"P(Y > {chi2_upper_val:.4f}) = {prob_upper:.6f}")
print("-" * 30)
print(f"Probabilidad total de rechazo: {total_prob:.6f}")
```

**Pregunta 3:** Si $S_{1}^{2}$ y $S_{2}^{2}$ son las desviaciones estándar de variables aleatorias independientes de tamaños $n_{1}=61$ y $n_{2}=31$ de poblaciones normales con $\sigma_{1}^{2}=12$ y $\sigma_{2}^{2}=18$ encuentre $P(S_{1}^{2}/S_{2}^{2})>1.16$

```python
# Datos del problema
n1, n2 = 61, 31
var1, var2 = 12, 18
ratio_s = 1.16

# Grados de libertad
df1 = n1 - 1
df2 = n2 - 1

# Calcular el valor F crítico a partir del cociente dado
f_value = ratio_s * (var2 / var1)

# Calcular la probabilidad P(F > f_value)
# Usamos la función de supervivencia (sf), que es 1 - cdf
prob_f = f.sf(f_value, dfn=df1, dfd=df2)

print(f"Grados de libertad: ν₁={df1}, ν₂={df2}")
print(f"Valor F calculado: {f_value:.4f}")
print("-" * 30)
print(f"La probabilidad P(S₁²/S₂² > 1.16) es: {prob_f:.4f}")
```

**Pregunta 4:** En la siguiente tabla para fumadores y no fumadores, compruebe si las varianzas poblacionales son iguales o no en $\alpha=0.05$:
- **Fumadores:** $n_{1}=26$, $S_{1}^{2}=36$
- **No fumadores:** $n_{2}=18$, $S_{2}^{2}=10$
1.  **Hipótesis:**
    * $H_0$: $\sigma_1^2 = \sigma_2^2$ (Las varianzas son iguales).
    * $H_1$: $\sigma_1^2 \neq \sigma_2^2$ (Las varianzas son diferentes).
2.  **Estadístico de prueba F:** $F = \frac{S_{1}^{2}}{S_{2}^{2}}$
3.  **Grados de libertad:** $\nu_1 = n_1 - 1 = 25$ y $\nu_2 = n_2 - 1 = 17$.
4.  **Regla de decisión:** Si p-value < $\alpha$ (0.05), se rechaza $H_0$.

```python
# Datos del problema
n1_fum, s1_sq = 26, 36
n2_nofum, s2_sq = 18, 10
alpha = 0.05

# Grados de libertad
df1_fum = n1_fum - 1
df2_nofum = n2_nofum - 1

# Calcular el estadístico F de prueba
f_test_stat = s1_sq / s2_sq

# Calcular el p-value para una prueba de dos colas
# Como F > 1, calculamos el área de la cola derecha y la duplicamos
p_value = 2 * f.sf(f_test_stat, dfn=df1_fum, dfd=df2_nofum)

print(f"Grados de libertad: ν₁={df1_fum}, ν₂={df2_nofum}")
print(f"Estadístico F de prueba: {f_test_stat:.4f}")
print(f"Nivel de significancia (α): {alpha}")
print(f"P-value calculado: {p_value:.4f}")
print("-" * 30)

# Conclusión
if p_value < alpha:
    print(f"Conclusión: Como el p-value ({p_value:.4f}) es menor que α ({alpha}), se rechaza la hipótesis nula.")
else:
    print(f"Conclusión: Como el p-value ({p_value:.4f}) es mayor que α ({alpha}), no se rechaza la hipótesis nula.")
```
