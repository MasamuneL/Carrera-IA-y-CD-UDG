```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
#establecemos la funcion
def f(x):
    return ((6.0*x**2)+(7.0*x))**2
#Creamos el array
x_array = np.linspace(-5,5, 100)
#derivamos con numpy
dfdx_numerical = np.gradient(f(x_array), x_array)

plt.plot(f(x_array))
plt.plot(dfdx_numerical)
```
