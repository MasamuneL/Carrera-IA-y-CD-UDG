```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Podemos establecer funciones para crear y graficar funciones
def parabola(x,a):
    return a*x**2.0
x = np.linspace(-10, 10, 200)

plt.plot(x,parabola(x,5))
```

```python
def f(x):
    function = x**3.0 * np.sin(x)
    return function
def dxf(x):
    dxfunc = (3.0*x**2 * np.sin(x)) + (x**3.0 * np.cos(x))
    return dxfunc
plt.plot(x, f(x))
plt.plot(x, dxf(x))
```
