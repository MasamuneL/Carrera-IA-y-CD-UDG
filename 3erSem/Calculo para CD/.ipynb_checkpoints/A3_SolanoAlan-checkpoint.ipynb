**Matemáticas Aplicadas a Ciencia de Datos**

En este notebook, se comparará el cálculo de derivadas por métodos analíticos con métodos numéricos computacionales. El objetivo es reforzar el concepto de la derivada y su comportamiento a través de la visualización.

## 1. Derivación Analítica

A continuación, se muestra el procedimiento para calcular la derivada de cada una de las funciones de forma analítica.

### Función 1: $f(x) = \sin(x^2)$

Para derivar esta función, utilizamos la **regla de la cadena**. Sea $u = x^2$, entonces $f(u) = \sin(u)$.

La regla de la cadena establece que $\frac{df}{dx} = \frac{df}{du} \cdot \frac{du}{dx}$.

1. Calculamos $\frac{df}{du}$:
   $$ \frac{d}{du}(\sin(u)) = \cos(u) $$

2. Calculamos $\frac{du}{dx}$:
   $$ \frac{d}{dx}(x^2) = 2x $$

3. Multiplicamos ambos resultados y sustituimos $u$ de nuevo por $x^2$:
   $$ f'(x) = \cos(x^2) \cdot 2x = 2x\cos(x^2) $$

### Función 2: $f(x) = e^{-x^2}$

Nuevamente, utilizamos la **regla de la cadena**. Sea $u = -x^2$, entonces $f(u) = e^u$.

1. Calculamos $\frac{df}{du}$:
   $$ \frac{d}{du}(e^u) = e^u $$

2. Calculamos $\frac{du}{dx}$:
   $$ \frac{d}{dx}(-x^2) = -2x $$

3. Multiplicamos y sustituimos $u$:
   $$ f'(x) = e^{-x^2} \cdot (-2x) = -2xe^{-x^2} $$

### Función 3: $f(x) = x^3 \cos(x)$

Para esta función, aplicamos la **regla del producto**: $(g \cdot h)' = g' \cdot h + g \cdot h'$.

Sean $g(x) = x^3$ y $h(x) = \cos(x)$.

1. Calculamos $g'(x)$:
   $$ g'(x) = \frac{d}{dx}(x^3) = 3x^2 $$

2. Calculamos $h'(x)$:
   $$ h'(x) = \frac{d}{dx}(\cos(x)) = -\sin(x) $$

3. Aplicamos la fórmula:
   $$ f'(x) = (3x^2)(\cos(x)) + (x^3)(-\sin(x)) = 3x^2\cos(x) - x^3\sin(x) $$

## 2 y 3. Implementación y Graficación en Python

Ahora, implementaremos las funciones y sus derivadas (analítica y numérica) en Python y las graficaremos para compararlas.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos el intervalo y el número de puntos
x = np.linspace(-3, 3, 200)
dx = x[1] - x[0] # Calculamos el paso para la derivada numérica
```

### Gráfica para $f(x) = \sin(x^2)$

```python
# 1. Definición de la función y su derivada analítica
def f1(x):
    return np.sin(x**2)

def df1_analytical(x):
    return 2 * x * np.cos(x**2)

# 2. Cálculo de la derivada numérica
df1_numerical = np.gradient(f1(x), dx)

# 3. Graficación
plt.figure(figsize=(10, 6))
plt.plot(x, f1(x), label='$f(x) = sin(x^2)$', color='blue')
plt.plot(x, df1_analytical(x), label="Derivada Analítica: $f'(x) = 2xcos(x^2)$", color='green', linestyle='-')
plt.plot(x, df1_numerical, label='Derivada Numérica', color='red', linestyle='--')
plt.title('Función 1 y sus Derivadas')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

### Gráfica para $f(x) = e^{-x^2}$

```python
# 1. Definición de la función y su derivada analítica
def f2(x):
    return np.exp(-x**2)

def df2_analytical(x):
    return -2 * x * np.exp(-x**2)

# 2. Cálculo de la derivada numérica
df2_numerical = np.gradient(f2(x), dx)

# 3. Graficación
plt.figure(figsize=(10, 6))
plt.plot(x, f2(x), label='$f(x) = e^{-x^2}$', color='blue')
plt.plot(x, df2_analytical(x), label="Derivada Analítica: $f'(x) = -2xe^{-x^2}$", color='green', linestyle='-')
plt.plot(x, df2_numerical, label='Derivada Numérica', color='red', linestyle='--')
plt.title('Función 2 y sus Derivadas')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

### Gráfica para $f(x) = x^3 \cos(x)$

```python
# 1. Definición de la función y su derivada analítica
def f3(x):
    return (x**3) * np.cos(x)

def df3_analytical(x):
    return 3 * (x**2) * np.cos(x) - (x**3) * np.sin(x)

# 2. Cálculo de la derivada numérica
df3_numerical = np.gradient(f3(x), dx)

# 3. Graficación
plt.figure(figsize=(10, 6))
plt.plot(x, f3(x), label='$f(x) = x^3 \cos(x)$', color='blue')
plt.plot(x, df3_analytical(x), label="Derivada Analítica", color='green', linestyle='-')
plt.plot(x, df3_numerical, label='Derivada Numérica', color='red', linestyle='--')
plt.title('Función 3 y sus Derivadas')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

## 4. Observaciones y Conclusiones

**¿Coinciden ambas curvas (analítica y numérica)?**

Sí, en todas las gráficas se puede observar que la curva de la derivada calculada de forma analítica (línea verde continua) y la derivada calculada de forma numérica (línea roja discontinua) son prácticamente idénticas. Se superponen casi a la perfección, lo que indica que el método numérico `numpy.gradient` es muy preciso para el intervalo y número de puntos seleccionados.

**¿En qué intervalos hay mayor o menor diferencia?**

Las diferencias, aunque mínimas, tienden a ser ligeramente mayores en los extremos del intervalo (cerca de -3 y 3). Esto es común en los métodos de diferencias finitas (como el que usa `gradient`), ya que en los bordes tienen menos información para calcular la pendiente. También pueden existir diferencias minúsculas en puntos donde la curvatura de la función cambia muy rápidamente. Sin embargo, visualmente, las diferencias son casi imperceptibles.

**¿Qué ventajas o desventajas observas entre los métodos?**

* **Método Analítico:**
    * **Ventajas:** Es exacto y proporciona una fórmula general para la derivada, que puede ser utilizada para análisis más profundos (como encontrar máximos y mínimos de forma exacta). No depende de la resolución de los puntos.
    * **Desventajas:** Puede ser muy difícil o incluso imposible de calcular para funciones muy complejas. Requiere conocimiento de las reglas de derivación y puede ser propenso a errores humanos.

* **Método Numérico:**
    * **Ventajas:** Es muy fácil y rápido de implementar computacionalmente, sin importar la complejidad de la función. Solo se necesita tener una serie de puntos de la función. Es extremadamente útil cuando no se tiene una expresión simbólica de la función, sino solo datos medidos.
    * **Desventajas:** Es una aproximación. Su precisión depende del espaciado entre los puntos (`dx`); un espaciado muy grande puede llevar a errores significativos. No proporciona una función simbólica como resultado, solo un conjunto de valores numéricos.
