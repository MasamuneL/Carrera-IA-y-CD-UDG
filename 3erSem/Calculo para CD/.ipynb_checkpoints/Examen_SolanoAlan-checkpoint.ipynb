## Sección 2 - Derivación simbólica

#### Objetivo

Aplicar herramientas de cálculo simbólico para
* Obtener derivadas de funciones algebraicas
* Identificar y clasificar puntos críticos
* Visualizar el comportamiento de la función y su derivada

Dada la siguiente función cúbica:

\begin{equation}
f(x) = x^3 - 6x^2 + 9x + 15
\end{equation}

1. **Define la función simbólicamente** usando `sympy`.

2. **Obtén la derivada primera** \\( f'(x) \\) de manera simbólica.

3. **Encuentra los puntos críticos** resolviendo la ecuación \\( f'(x) = 0 \\).

4. **Clasifica los puntos críticos** usando la derivada segunda. Indica si se trata de un **máximo local**, **mínimo local** o **punto de inflexión**.

5. **Grafica la función y su derivada** en el intervalo \\( x \in [-15, 15] \\). Asegúrate de etiquetar el gráfico y usar líneas de distinto estilo para identificar sus puntos críticos.

### Respuestas Seccion #2

1. **Define la función simbólicamente** usando `sympy`.

````python
## Ejemplo de código
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Declarar la variable simbólica
x = sp.symbols('x')

# Definir la función 
f = x**3 + 6*x**2 + 9*x + 15
print('Función: ')
display(f)
````

2. **Obtén la derivada primera** \\( f'(x) \\) de manera simbólica.

````python
# Primera derivada
dfdx = sp.diff(f,x)
print('Primera derivada: ')
display(dfdx)
````

3. **Encuentra los puntos críticos** resolviendo la ecuación \\( f'(x) = 0 \\).

````python
# Puntos críticos
crits = sp.solve(dfdx, x)
print('Puntos críticos: ')
display(crits)
````

4. **Clasifica los puntos críticos** usando la derivada segunda. Indica si se trata de un **máximo local**, **mínimo local** o **punto de inflexión**.

````python
# Segunda derivada para ver si estos puntos críticos son máximo, mínimo o punto de inflexion
d2fdx2 = sp.diff(dfdx, x)
print('Clasificación de puntos críticos')
for punto in crits:
    segunda = d2fdx2.subs(x, punto)
    tipo = "mínimo" if segunda > 0 else "máximo" if segunda < 0 else "punto de inflexión"
    print(f"x = {punto}, f''(x) = {segunda} → {tipo}")
````

### Respuestas Seccion #3

1. Deriva simbólicamente la función y convierte su derivada en una función de Python usando `lambdify`.

````python
# Código de ejemplo
import sympy as sp

# Definir la función simbólicamente
x = sp.symbols('x')
f = x**3 - 6*x**2 + 9*x + 15

# Derivada simbólica
dfdx_sym = sp.diff(f, x)
dfdx = sp.lambdify(x, dfdx_sym)

display(f) # Mostrar función original
display(dfdx_sym)  # Mostrar derivada simbólica
````

2. Utiliza la función `gradient_descent` proporcionada para encontrar el mínimo local, comenzando desde $x_0 = 4.0$, con una tasa de aprendizaje de 0.01 y 50 iteraciones.

````python
def gradient_descent(dfdx, x, learning_rate=0.1, num_iterations=100):
    for iteration in range(num_iterations):
        x = x - learning_rate * dfdx(x)
    return x
````

````python
# Aplicación de gradiente descendente
x_min = gradient_descent(dfdx, x=4.0, learning_rate=0.1, num_iterations=50)
print("Valor de x donde la función f es mínimo: ", x_min)
````

3. Modifica la función e imprime los últimos 5 valores obtenidos por el algoritmo. ¿Ves que los valores se estabilizan? ¿Qué indica eso?

````python
def gradient_descent_last_5(dfdx, x, learning_rate=0.1, num_iterations=100):
    resultados = []
    ultimos5 = num_iterations - 5
    for iteration in range(num_iterations):
        x = x - learning_rate * dfdx(x)
        resultados.append(x)
    return resultados[ultimos5:]
````

````python
x_min_last5 = gradient_descent_last_5(dfdx, x=4.0, learning_rate=0.1, num_iterations=50)
print("Valor de x donde la función f es mínimo: ", x_min_last5)
````

- Significa que se esta cercando a u punto minimo ya que no hay tanta variacion en los valores o en mi caso que ya llego al valor minimo

4. Cambia el punto inicial a $x_0 = -1.0$. ¿A qué mínimo llega ahora? ¿Qué puedes concluir sobre los **mínimos locales**?

````python
x_min_last5_2 = gradient_descent_last_5(dfdx, x=-1.0, learning_rate=0.003, num_iterations=50)
print("Valor de x donde la función f es mínimo: ", x_min_last5_2)
````

- Se pierde en el gradiente ya que la bajada es muy amplia por lo cual no hay convergencia
