```python
# Importación de librerías
import numpy as np
import matplotlib.pyplot as plt
```

## Representación de un sistema de ecuaciones lineales utilizando matrices

Supongamos que tenemos el siguiente **sistema lineal de ecuaciones**, compuesto de tres ecuaciones y tres incógnitas,

\begin{equation}
\begin{cases}
4x_1 - 3x_2 + x_3 = -10 \\
2x_1 + x_2 + 3x_3 = 0 \\
-x_1+ 2x_2 -5x_3 = 17
\end{cases}
\end{equation}

Para resolver este sistema de ecuaciones necesitamos encontrar valores de $x_1, x_2, x_3$ que sean una solución simultánea de las tres ecuaciones anteriores.

### Resolviendo este sistema de ecuaciones utilizando representación matricial

Para esto, vamos a utilizar la librería ``numpy``. $\mathbf{A}$ será una matriz donde cada fila represente una ecuacion en el sistema y cada columna corresponde a una variable $x_1, x_2, x_3$. $\mathbf{b}$ será un arreglo 1-dimensional que represente a los coeficientes constantes del lado derecho de las igualdades.

```python
A = np.array([[4,-3,1],
             [2,1,3],
             [-1,2,-5]],
             dtype = np.dtype(float))

b = np.array([-10,0,17],
             dtype = np.dtype(float))

print("Matriz A: ")
print(A)
print("Array b: ")
print(b)
```

```python
# Checamos las dimensiones de A y b
print(f'Shape de matriz A: {A.shape}')
print(f'Shape de array b: {b.shape}')
```

Rápidamente, podemos utilizar el método `np.linalg.solve(A,b)` para encontrar la solución a este sistema. El resultado será un arreglo 1-dimensional, donde cada elemento es una solución a cada variable $x_1, x_2, x_3$.

```python
x = np.linalg.solve(A,b)
print(f'Soluciones: {x}')
```

## Evaluando el determinante de una matriz

La matriz $\mathbf{A}$ es una **matriz cuadrada**, y tiene el mismo número de filas que de columnas. En este caso, es posible calcular un determinante - un número real que caracteriza algunas propiedades de esta matriz. Un sistema lineal de ecuaciones tendrá una solución sí y solo sí la matriz $\mathcal{A}$ tiene un determinante no nulo. 

El determinante lo podemos calcular utilizando el método ``np.linalg.det()``.

```python
d = np.linalg.det(A)
print(f'Determinante de A: {d:.2f}')
```

Entonces, hay librerías que nos ayudan a resolver rápidamente sistemas lineales de ecuaciones. Sin embargo, para entender más profundamenet el tema y sus conceptos matemáticos, es importante practicar con técnicas manuales de resolución.

## Resolviendo un sistema de ecuaciones con método de Gauss-Jordan

Resolver sistemas de ecuaciones lineales es una de las tareas fundamentales en álgebra lineal. Cuando se trabaja con tres variables o más, uno de los métodos más sistemáticos y potentes para encontrar la solución exacta es el **método de Gauss-Jordan**, una extensión del clásico **método de eliminación de Gauss**.

Ambos métodos se basan en transformar la **matriz aumentada** del sistema mediante **operaciones elementales por fila**:
- Intercambiar filas
- Multiplicar una fila por un escalar no nulo
- Sumar o restar un múltiplo de una fila a otra

### Método de Gauss (o eliminación de Gauss)

Este método busca llevar la matriz aumentada a una **forma escalonada** (Row Echelon Form, REF). En esta forma:
- Todos los coeficientes por debajo del pivote (el primer número no nulo de una fila) son cero.
- Los pivotes se mueven progresivamente a la derecha en cada fila inferior.
- Las filas de ceros (si las hay) se colocan al final.

Una vez que la matriz está en forma escalonada, se utiliza **sustitución hacia atrás** (back-substitution) para resolver el sistema: se despeja primero la última variable y se sustituyen sus valores en las ecuaciones anteriores hasta obtener todas las incógnitas.

**Ventaja**: Menor número de operaciones que Gauss-Jordan.  
**Desventaja**: Requiere una fase adicional de sustitución hacia atrás.

### Método de Gauss-Jordan

En cambio, el método de Gauss-Jordan lleva la matriz hasta su forma **escalonada reducida por filas** (Row-Reduced Echelon Form, RREF). Además de lo logrado por el método de Gauss, ahora también se asegura que:
- Todos los pivotes sean igual a 1 (si no lo eran).
- Todos los elementos por **encima y por debajo** de cada pivote sean cero.

Como resultado, no es necesario hacer sustitución hacia atrás: **la solución puede leerse directamente** desde la matriz resultante.

**Ventaja**: Resultado inmediato sin necesidad de retro-sustitución.  
**Desventaja**: Implica más operaciones (menos eficiente para sistemas grandes).

### Comparación rápida

| Característica                     | Eliminación de Gauss               | Gauss-Jordan                         |
|------------------------------------|------------------------------------|--------------------------------------|
| Forma final                        | Escalonada (REF)                   | Escalonada reducida (RREF)           |
| ¿Requiere sustitución hacia atrás? | Sí                                 | No                                   |
| Operaciones sobre filas            | Sí                                 | Sí                                   |
| Complejidad computacional          | Menor                              | Mayor (más operaciones)              |
| Uso típico                         | Más eficiente en sistemas grandes  | Más didáctico para sistemas pequeños |

Ambos métodos comparten el mismo espíritu: simplificar el sistema utilizando transformaciones equivalentes. Sin embargo, el **método de Gauss-Jordan** resulta más transparente para visualizar directamente la solución, por lo que es muy utilizado con fines educativos y en casos donde se requiere la forma canónica de una matriz (por ejemplo, para calcular la inversa).

## Preparación para la reducción de las filas

Para empezar, vamos a unificar la matriz $\mathcal{A}$ con el array $\mathcal{b}$ mediante la operación ``np.hstack()``. Para hacer esto es necesario reestructurar las dimensiones del array $\mathcal{b}$, para que coincidan,

```python
A_system = np.hstack((A, b.reshape((3,1))))
print(A_system)
```

Para facilitar el proceso, vamos a definir funciones para operaciones elementales, que no cambian la solución del sistema lineal de ecuaciones:

* Multiplicar una fila por un número real (diferente de cero)
* Sumar dos filas e intercambiar una de las filas originales con el resultado de la suma
* Intercambiar filas

```python
def MultiplyRow(M, row_num, row_num_multiple):
    '''
    Multiplica una fila especifica de una matriz por un escalar
    '''
    M_new = M.copy() # Hacer una copia es requerido para no alterar el objeto inicial
    M_new[row_num] = M_new[row_num] * row_num_multiple

    return M_new

print(f'Matriz original: \n {A_system}')
print(f'Matriz después de multiplicar su tercer fila por 2: \n {MultiplyRow(A_system, 2, 2)}')

    
```

```python
def AddRows(M, row_num_1, row_num_2, row_num_1_multiple):
    '''
    Multiplica una fila de la matriz por un escalar y la suma a otra fila
    '''
    M_new = M.copy()
    M_new[row_num_2] = row_num_1_multiple * M_new[row_num_1] + M_new[row_num_2]
    return M_new

print(f'Matriz original: \n {A_system}')
print(f'Matriz después de intercambiar la tercer fila con la suma de si misma y la segunda multiplicada por 1/2: \n {AddRows(A_system,1,2,1/2)}')
```

```python
def SwapRows(M,row_num_1, row_num_2):
    '''
    Intercambia dos filas de una matriz
    '''
    M_new = M.copy()
    M_new[[row_num_1, row_num_2]] = M_new[[row_num_2, row_num_1]]
    return M_new

print(f'Matriz original: \n {A_system}')
print(f'Matriz después de intercambiar la primer y tercera fila: \n {SwapRows(A_system,0,2)}')
```

### Implementación del método de Gauss Jordan

```python
def GaussJordan(M):
    '''
    Aplica el método de Gauss-Jordan a una matriz aumentada para resolver un sistema de ecuaciones lineales
    '''
    M_new = M.copy().astype(float)
    n_rows, n_cols = M_new.shape

    # Revisar que el determinante no sea igual a cero
    coef = M_new[:, :-1] # submatriz de coeficientes
    if abs(np.linalg.det(coef)) < 1e-10:
        raise ValueError("El sistema no tiene solución única (determinante ≈ 0).")

    print("Matriz inicial:")
    print(M_new, "\n")

    # Algoritmo para Gauss-Jordan
    # Nos aseguramos que el pivote no sea cero, y si es así, intercambiamos filas con uno que no lo sea
    for i in range(n_rows):
        if abs(M_new[i][i]) < 1e-10:
            for k in range(i+1,n_rows):
                if abs(M_new[k][i]) > 1e-10:
                    M_new = SwapRows(M_new, i, k)
                    print(f"Intercambio de fila {i} con fila {k}:")
                    print(M_new, "\n")
                    break

        # normalizar la fila i para que el pivote sea 1
        pivot = M_new[i, i]
        if abs(pivot) > 1e-10:
            M_new = MultiplyRow(M_new, i, 1/pivot)
            print(f"Normalización de la fila {i} (pivote → 1):")
            print(M_new, "\n")

        # hacer ceros en la columna i para todas las demás filas
        for j in range(n_rows):
            if j != i:
                factor = -M_new[j, i]
                if abs(factor) > 1e-10:
                    M_new = AddRows(M_new, i, j, factor)
                    print(f"Fila {j} → Fila {j} + ({factor:.3f}) * Fila {i}:")
                    print(M_new, "\n")

    print("Matriz final en forma reducida por filas (RREF):")
    print(M_new)
    
    return M_new, M_new[:, -1]
```

```python
GaussJordan(A_system)
```

### Sistema de ecuaciones lineales sin solución

\begin{equation}
\begin{cases}
x_1 + x_2 + x_3 = 2 \\
x_2 - 3x_3 = 1 \\
2x_1 + x_2 + 5x_3 = 0
\end{cases}
\end{equation}

```python
A_2= np.array([
        [1, 1, 1],
        [0, 1, -3],
        [2, 1, 5]
    ], dtype=np.dtype(float))

b_2 = np.array([2, 1, 0], dtype=np.dtype(float))

A_2_system = np.hstack((A_2, b_2.reshape((3, 1))))
```

```python
GaussJordan(A_2_system)
```
