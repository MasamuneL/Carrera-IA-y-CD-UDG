```python
# evaluarCadena reporta si la cadena es aceptada o no.
def evaluarCadena(cadena): {
	
	# Convertir la cadena a minúsculas y verificar que solo contenga "a" y "b".
	cadena = cadena.toLowerCase();
	if (!/^[ab]*$/.test(cadena)) {
		console.log(`\x1b[31mError:\x1b[0m La cadena contiene caracteres no permitidos. Solo se aceptan "a" y "b".`);
		return undefined;
	}

	# ================================================================ #
	# ========== AUTÓMATA ============================================ #

	let estado = 0; // Estado actual, comienza en estado inicial (0).
	
    # Para cada caracter de la cadena... 
	for (let i = 0; i < cadena.length; i++) {
		
		let simbolo = cadena[i]; // Caracter actual

		if (estado === 0) {
			if (simbolo === 'a') {
				estado = 1;
			} else {
				estado = 2;
			}

		} else if (estado === 1) {
			if (simbolo === 'b') {
				estado = 2;
			} else {
				estado = 3;
			}

		} else if (estado === 2) {
			if (simbolo === 'b') {
				estado = 3;
			} else {
				estado = 0;
			}

		} else if (estado === 3) {
			if (simbolo === 'a') {
				estado = 4;
			} else {
				estado = 4;
			}

		} else {
			# Terminar con mensaje de error para estados inválidos.
			let cadenaColoreada = cadena.split('').map((c, index) => index === i ? `\x1b[31m${c}\x1b[0m` : c).join('');
			console.log(`\x1b[31mError:\x1b[0m El estado "\x1b[33m${estado}\x1b[0m" no existe. Evaluando símbolo #${i + 1} de cadena "${cadenaColoreada}".`);
			return undefined
		}
	}

	# ================================================================ #
	# ========== ESTADO DE ACEPTACIÓN ================================ #
	
	# Si tienes más de uno hay que agregarlos.
	if (estado === 4) {
		return true
	} else {
		return false
	}
}
```
