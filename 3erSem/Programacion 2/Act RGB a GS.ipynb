```python
#Creamos una funcion para imprimir en rgb
def print_rgb(text, r, g, b):
    rgb_code = f"\033[38;2;{r};{g};{b}m"
    reset_code = "\033[0m"
    print(f"{rgb_code}{text}{reset_code}")

#Debemos crear varibles para asignar los valores de RGB
r = 0
g = 0
b = 0

#Para hacerlo mas dinamico podemos hacer que estos valores sean elegidos por el usuario mediante un input, agramos un try por si el usuario se equivoca de valor
try:
    r = round(int(input("Ingrese la cadntidad de Rojo que quiere que tenga su color (max. 255)")))
    g = round(int(input("Ingrese la cadntidad de Verde que quiere que tenga su color (max. 255)")))
    b = round(int(input("Ingrese la cadntidad de Azul que quiere que tenga su color (max. 255)")))

    #Aseguramos el rango correcto
    r = max(0, min(255, r))
    g = max(0, min(255, g))
    b = max(0, min(255, b))
    
except ValueError:
    print("Error Tipo de valor no valido")

#Muestra los valores input
print(f'Rojo = {r}, Verde = {g}, Azul = {b}')
print_rgb("¡Tu color!", r, g, b)
```

```python
#Ya con esto podemos empezar a crear el programa que se encargara de transformarlo a escala de grises
#Descartamos que todos los valores sean iguales
if r == g and r == b:
    print("tu color ya esta en escala de grises ;)")
else:
    #Usamos la formula de luminicidad para crear los colores a grises
    gs_value = round(r*0.299 + g*0.587 + b*0.114)
    print(f'Segun la formula de luminicidad tu nuevo color seria: R = {gs_value}, G = {gs_value}, B = {gs_value}')
    print_rgb("¡Tu color en escala de gris!", gs_value, gs_value, gs_value)
```
